# Phase 1: Data Ingestion

## Set up

```python
import pandas as pd
from pathlib import Path
import requests
import json

# The website where the data is located
DATA_SOURCE_URL = "https://gist.githubusercontent.com/tkfu/9819e4ac6d529e225e9fc58b358c3479/raw/d4df8804c25a662efc42936db60cfbc0a5b19db8/srd_5e_monsters.json"

# The relative path on the user's local machine where the data will be stored
CURR_DIR = Path(".").resolve().parent #returns an absolute path to the current working directory, no needed to use Path.resolve(). Alt: Path(".")

DATA_DIR = CURR_DIR / "data" / "raw" #File structure that houses monster.json

LOCAL_DATA_PATH = DATA_DIR / "monsters_raw.json"  #The full file path. Used to check if it exists in the first place

pd.set_option("display.max_columns", None)
```

## Caching check
* Does a cache check to ensure the folder structure and dataset is on the local device. If its not it grabs it from the website, builds the folder structure and then downloads the datset onto the user's device

```python
if not LOCAL_DATA_PATH.exists():
    print("File Path not found! Creating it now...")
    DATA_DIR.mkdir(parents=True, exist_ok=True) # Makes the directory on the user's local machine. If its already there or needs parents then it handles it
    r = requests.get(DATA_SOURCE_URL) # Gets the dataset from the Github website
    with open(LOCAL_DATA_PATH, "w") as f:
        f.write(r.text) #caches the data set onto the user's local machine 

else: 
    print("File path validated! The Raw dataset is on your computer!")
# Job done. Following code just to see if worked
with open(LOCAL_DATA_PATH, "r") as f:
    json_obj = json.load(f) #loads the json file into a python object

raw_df = pd.json_normalize(json_obj)
raw_df.head()

```
